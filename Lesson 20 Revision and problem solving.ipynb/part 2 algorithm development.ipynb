```python
def almeer_min(x):
    y = 1000000000000000000000
    for each in x:
        if each < y:
            y = each
    return y
```

```python
def add_sum(x):
    y = 0
    for each in x:
        y = y + each
    return y
```

```python
def add_avg(x):
    y = 0
    c =0 
    for each in x:
        y = y + each
        c = c + 1
    return y/c
```
