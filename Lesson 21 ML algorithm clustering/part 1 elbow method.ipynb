```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

def kmeans_clustering(data, centroids):
    kmeans = KMeans(n_clusters=len(centroids), init=centroids, n_init=1)
    labels = kmeans.fit_predict(data)
    return labels

def plot_clusters(data, labels, centroids):
    plt.figure(figsize=(8, 6))
    colors = ['r', 'g', 'b', 'y', 'c', 'm']
    
    for i in range(len(centroids)):
        cluster_data = data[labels == i]
        plt.scatter(cluster_data[:, 0], cluster_data[:, 1], c=colors[i], label=f'Cluster {i+1}')
    
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, c='k', label='Centroids')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('KMeans Clustering')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
data = np.random.rand(100, 2)  # Sample data
centroids = np.random.rand(3, 2)  # Sample centroids

labels = kmeans_clustering(data, centroids)
plot_clusters(data, labels, centroids)
```
